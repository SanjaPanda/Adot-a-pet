```python
import pandas as pd
import numpy as np 
from datetime import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
```

## Initial data cleaning

```python
pet_df = pd.read_csv('Data/clean_train.csv')
pet_df.head()
```

```python
pet_df_test = pd.read_csv('Data/clean_test.csv')
pet_df_test.head()
```

```python
pet_df.info()
```

```python
pet_df['issue_date'] = pd.to_datetime(pet_df['issue_date']) 
pet_df['listing_date'] = pd.to_datetime(pet_df['listing_date']) 

pet_df_test['issue_date'] = pd.to_datetime(pet_df_test['issue_date']) 
pet_df_test['listing_date'] = pd.to_datetime(pet_df_test['listing_date']) 
```

```python
# categorical feature
pet_df['breed_category'] = pet_df['breed_category'].astype('object')
pet_df['pet_category'] = pet_df['pet_category'].astype('object')
pet_df['condition'] = pet_df['condition'].astype('object')
pet_df['listing_year'] = pet_df['listing_year'].astype('object')
pet_df['issue_year'] = pet_df['issue_year'].astype('object')
pet_df['listing_month'] = pet_df['listing_month'].astype('object')
pet_df['issue_month'] = pet_df['issue_month'].astype('object')
```

```python
pet_df.info()
```
